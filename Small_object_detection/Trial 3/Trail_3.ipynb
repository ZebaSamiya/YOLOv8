## Loading YOLOv8 Model

```python
from ultralytics import YOLO
model = YOLO("yolov8n.pt")
```

```python
model.info()
```

## Training Model

```python
results = model.train(
    data="traffic-sign-detection-yolov8-10/data.yaml",
    epochs=40,
    imgsz=1024,
    batch=16,
    optimizer='AdamW',
    lr0=0.005,
    lrf=0.0002,
    momentum=0.93,
    weight_decay=0.0005,
    iou=0.45,
    conf=0.4,
    mosaic=0.8,
    scale=0.4,
    flipud=0.4,
    fliplr=0.5,
    degrees=5,
    save=True
)
```

## Running Inference

```python
import os
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Enable inline plotting for Jupyter
%matplotlib inline  

# Define paths
test_images_folder = "./traffic-sign-detection-yolov8-10/test/images"
output_folder = "./detections"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Load YOLO model
model = YOLO("runs/detect/train6/weights/best.pt")  # Ensure correct model file

# Run inference on all images
for image_name in sorted(os.listdir(test_images_folder)):
    if image_name.endswith((".jpg", ".png", ".jpeg")):
        image_path = os.path.join(test_images_folder, image_name)

        # Run inference
        results = model(image_path, conf=0.5, iou=0.4)

        # Save detected image
        for result in results:
            output_image = result.plot()
            output_path = os.path.join(output_folder, f"detected_{image_name}")
            cv2.imwrite(output_path, output_image)

        # print(f"Processed: {image_name} -> Saved to {output_path}")

print("Inference completed on all test images!")

# Function to display detected images
def show_detected_images(folder_path, num_images=16):
    image_files = sorted(os.listdir(folder_path))[:num_images]

    if not image_files:
        print("No detected images found!")
        return

    rows, cols = 4, 4
    fig, axes = plt.subplots(rows, cols, figsize=(15, 15))

    for i, img_file in enumerate(image_files):
        img_path = os.path.join(folder_path, img_file)
        print(f"Loading: {img_path}")  # Debugging step
        img = cv2.imread(img_path)

        if img is None:
            print(f" Warning: Failed to load {img_file}")
            continue

        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        row, col = divmod(i, cols)
        axes[row, col].imshow(img)
        axes[row, col].axis("off")
        axes[row, col].set_title(img_file, fontsize=10)

    plt.tight_layout()
    plt.show()

# Show images
show_detected_images(output_folder, num_images=16)
```
